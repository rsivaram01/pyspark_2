```python
f_df  = spark.read.options(header = 'True', inferSchema = 'True').csv("/Volumes/org_data/emp_data/practice_volume_1/csv_data/atc_pre_departure_delays_2016 (1).csv")
```

```python
f_df.count()
```

```python
# while reading file it self spark will partiton the data
```

```python
from pyspark.sql.functions import *
```

```python
sp_df = f_df.withColumn("s_pid", spark_partition_id())
```

```python
sp_df.groupBy('s_pid').count().display()
```

```python
to get benifit out of parllsim

Data skewness ===> getting one pratiton with high records , other partition will be having less number of records 
```

```python
sp_df1 = sp_df.coalesce(3)
```

```python
sp_df1.withColumn('sp_id2', spark_partition_id())\
    .groupBy('sp_id2').count()\
     .display()
```

```python
# reparttion ===> it is used to increase partition and coalesce is used to decrease the partition

sp_df1.repartition(10).withColumn('sp_id2', spark_partition_id())\
    .groupBy('sp_id2').count()\
     .display()
```

```python
final_df = sp_df1.repartition(10)
```

```python
final_df.coalesce(1).write.mode("overwrite").format('csv').options(header=True).save("/Volumes/org_data/emp_data/practice_volume_1/write_csv/")
```
