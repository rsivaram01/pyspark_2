```python
from pyspark.sql.functions import *
```

```python
emp_df1 = spark.read.options(header='true', inferSchema='true').csv("/Volumes/sample_cat/default/volume_csv/emp_dataset/employees.csv")
```

```python
emp_df1.withColumn("s_pid", spark_partition_id()).groupBy('s_pid').count().display()
```

```python
emp_df2 = emp_df1.filter((col("SALARY")>10000) & (month(to_date(col("HIRE_DATE"), 'dd-MMM-yy')) == 6))

emp_df3 = emp_df2.withColumn("full_name",concat(col("FIRST_NAME"), lit(" "), col("LAST_NAME")))


emp_df3.select('EMPLOYEE_ID',"full_name").write.mode("overwrite").saveAsTable("org_data.emp_data.emp")
```
